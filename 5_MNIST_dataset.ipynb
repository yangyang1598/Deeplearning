## 1. 데이터 불러오기

## 2. 데이터 시각화 (EDA)

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,9))
for i in range(30):
  plt.subplot(5,6,i+1)
  img=x_train[i]
  label=y_train[i]
  plt.title(label)
  plt.imshow(img)
  plt.xticks([])
  plt.yticks([])
plt.show()
```
